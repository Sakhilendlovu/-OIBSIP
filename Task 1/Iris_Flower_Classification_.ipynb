Creating the scatter plot

```python
# Create a scatter plot
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
# Set labels and legend
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)

# Show the plot
plt.show()
```

Checking all Null Values in Dataset


```python
# prompt: check Null values in dataset

# Load Iris dataset
iris = load_iris()

# Create a DataFrame from the data and feature names
Iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Drop any rows with missing values
Iris_df.dropna(inplace=True)

# Check for any missing values
print("Missing Values:\n", Iris_df.isnull().sum())

# Show the first few rows of the DataFrame
print("\nFirst Few Rows of the DataFrame:\n", Iris_df.head())
```

Visualisation

```python
# Load Iris dataset
iris = load_iris()

# Create a DataFrame from the data and feature names
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Visualize the data
sns.pairplot(iris_df, hue="species", markers="x")
plt.show()
```
