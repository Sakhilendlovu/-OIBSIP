```python
import pandas as pd
import plotly.graph_objects as go
df_11_2020.columns
df_11_2020 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
fig = go.Figure()

fig.add_traces(
    go.Bar(x=df_11_2020, y=df_11_2020['Estimated Unemployment Rate'], name='Unemployment Rate'),
    go.Bar(x=df_11_2020, y=df_11_2020['Estimated Labour Participation Rate'], name='Labour Participation Rate')
)

fig.update_layout(
    title='Unemployment Rate and Labour Participation Rate (Nov 2020)',
    xaxis={
        'categoryorder': 'array',
        'categoryarray': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    }
)

fig.show()
```
